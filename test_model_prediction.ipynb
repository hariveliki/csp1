```python
import torch
import torch.nn as nn
from resnet import ResNet
from convnet import ConvNet
from torchsummary import summary
```

```python
wp = ResNet()
boiler = ConvNet()
wm = ConvNet()
```

```python
wp.load_state_dict(
    torch.load(
        "result/wärmepumpe/ResNet_0/ResNet_0.pt", map_location=torch.device("mps")
    )["best_model_state_dict"]
)
boiler.load_state_dict(
    torch.load(
        "result/boiler/ConvNet_4/ConvNet_4.pt", map_location=torch.device("mps")
    )["best_model_state_dict"]
)
wm.load_state_dict(
    torch.load(
        "result/waschmaschine/ConvNet_1/ConvNet_1.pt", map_location=torch.device("mps")
    )["best_model_state_dict"]
)
```

```python
models = {"wp": wp, "boiler": boiler, "wm": wm}
for model in models.values():
    model.eval()
```

```python
import pandas as pd
xtest = pd.read_csv("data/inference/xtest.csv")
xtest.iloc[:, 2:]
```

```python
import numpy as np
import torch

xtest_values = xtest.iloc[:,2:].values.astype(np.float32)
t = torch.tensor(xtest_values)
t = t.unsqueeze(0)
t[:, 1:2, :].shape
```

```python
result = {"wp": 0, "boiler": 0, "wm": 0}
with torch.no_grad():
    for key, model in models.items():
        # BATCH_SIZE x SMART_METER x TS_LENGTH
        logits = model(t[:, 1:2, :])
        pred = torch.argmax(logits, dim=1)
        result[key] = pred.item()
result
```
